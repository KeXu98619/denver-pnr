```python
tracts_pop = pd.read_csv(r".\Data\ACSDP5Y2024.DP05_2026-02-26T152902\ACSDP5Y2024.DP05-Data.csv")
```

```python
tracts_pop = tracts_pop.loc[1:]
tracts_pop = tracts_pop[['GEO_ID','DP05_0001E']].rename(columns = {'DP05_0001E':'pop'})
```

```python
# ── Fix GEO_ID format and merge population into tracts ───────────────────────
# tracts_pop GEO_ID looks like "1400000US08001007801"
# tracts GEOID looks like           "08001007801"
tracts_pop['GEOID'] = tracts_pop['GEO_ID'].str.replace('1400000US', '', regex=False)
tracts_pop['pop']   = pd.to_numeric(tracts_pop['pop'], errors='coerce').fillna(0)

tracts = tracts.merge(tracts_pop[['GEOID', 'pop']], on='GEOID', how='left')
tracts['pop'] = tracts['pop'].fillna(0)

print(f"Tracts total: {len(tracts)}")
print(f"Tracts with population > 0: {(tracts['pop'] > 0).sum()}")
tracts[['GEOID', 'pop']].head()
```

```python
# ── Separate region GDF into PnR locations vs hex regions ────────────────────
# Regions are named "Region XXX"; PnR locations have real street names
pnr_gdf    = region[~region['name'].str.startswith('Region')].copy()
region_gdf = region[ region['name'].str.startswith('Region')].copy()

pnr_names = set(pnr_gdf['name'])

print(f"PnR locations : {len(pnr_gdf)}")
print(f"Hex regions   : {len(region_gdf)}")

# ── Filter OD data ────────────────────────────────────────────────────────────
# Drop External on either side
od_clean = od[
    (od['Origin']      != 'External') &
    (od['Destination'] != 'External')
].copy()

od_clean['orig_is_pnr'] = ~od_clean['Origin'].str.startswith('Region')
od_clean['dest_is_pnr'] = ~od_clean['Destination'].str.startswith('Region')

# Keep only PnR→Region and Region→PnR  (drop PnR→PnR and Region→Region)
pnr_to_region = od_clean[ od_clean['orig_is_pnr'] & ~od_clean['dest_is_pnr']][['Origin', 'Destination', 'trips']].copy()
region_to_pnr = od_clean[~od_clean['orig_is_pnr'] &  od_clean['dest_is_pnr']][['Origin', 'Destination', 'trips']].copy()

print(f"\nPnR→Region rows : {len(pnr_to_region):,}  |  trips : {pnr_to_region['trips'].sum():,.0f}")
print(f"Region→PnR rows : {len(region_to_pnr):,}  |  trips : {region_to_pnr['trips'].sum():,.0f}")
```

```python
# ── Reproject everything to EPSG 3857 (metres) for accurate area calculation ──
region_gdf = region_gdf.to_crs(3857)
tracts     = tracts.to_crs(3857)

tracts['tract_area'] = tracts.geometry.area   # m²

print(f"region_gdf CRS : {region_gdf.crs}")
print(f"tracts CRS     : {tracts.crs}")

# ── Spatial overlay: hex regions × census tracts ──────────────────────────────
intersect_gdf = gpd.overlay(
    region_gdf[['name', 'geometry']],
    tracts[['GEOID', 'pop', 'tract_area', 'geometry']],
    how='intersection'
)
intersect_gdf['intersect_area'] = intersect_gdf.geometry.area

print(f"\nIntersection pieces: {len(intersect_gdf):,}")
```

```python
# ── Compute region→tract weights ──────────────────────────────────────────────
#
# Logic (mirrors the example):
#   1. Each intersection piece gets a share of its tract's population
#      proportional to what fraction of the tract it covers:
#          pop_allocated = tract_pop × (intersect_area / tract_area)
#   2. For each hex region, sum those allocated pops to get the region total.
#   3. weight = pop_allocated / region_total_pop
#      → weights per region sum to 1 and tell us how to split trips to tracts.
#   4. Fallback to area-based weight for regions with zero population.

# Step 1 – allocate pop by area share within the tract
intersect_gdf['pop_allocated'] = (
    intersect_gdf['pop'] * (intersect_gdf['intersect_area'] / intersect_gdf['tract_area'])
)

# Step 2 – total allocated pop per region (denominator)
region_total_pop  = intersect_gdf.groupby('name')['pop_allocated'].sum().rename('region_total_pop')
region_total_area = intersect_gdf.groupby('name')['intersect_area'].sum().rename('region_total_area')

intersect_gdf = intersect_gdf.join(region_total_pop,  on='name')
intersect_gdf = intersect_gdf.join(region_total_area, on='name')

# Step 3 – weight (pop-based, area fallback for zero-pop regions)
intersect_gdf['pop_weight']  = intersect_gdf['pop_allocated']   / intersect_gdf['region_total_pop']
intersect_gdf['area_weight'] = intersect_gdf['intersect_area']  / intersect_gdf['region_total_area']
intersect_gdf['weight'] = np.where(
    intersect_gdf['region_total_pop'] > 0,
    intersect_gdf['pop_weight'],
    intersect_gdf['area_weight']
)

region_tract_weights = intersect_gdf[['name', 'GEOID', 'weight']].copy()

# Sanity check: weights should sum to ~1.0 per region
weight_sums = region_tract_weights.groupby('name')['weight'].sum()
print("Weight-sum stats per region (should all be ~1.0):")
print(weight_sums.describe().round(4))
region_tract_weights.head()
```

```python
# ── PnR → Region  →  PnR → Tract ─────────────────────────────────────────────
pnr_to_tract = (
    pnr_to_region
    .merge(region_tract_weights, left_on='Destination', right_on='name', how='left')
)

pnr_to_tract['trips_allocated'] = pnr_to_tract['trips'] * pnr_to_tract['weight']

pnr_to_tract = (
    pnr_to_tract
    .groupby(['Origin', 'GEOID'])['trips_allocated']
    .sum()
    .reset_index()
    .rename(columns={'Origin': 'pnr', 'GEOID': 'tract_id', 'trips_allocated': 'trips'})
)

print(f"PnR→Tract rows  : {len(pnr_to_tract):,}")
print(f"Trips in        : {pnr_to_region['trips'].sum():,.1f}")
print(f"Trips out       : {pnr_to_tract['trips'].sum():,.1f}  (difference due to regions outside tract coverage)")
pnr_to_tract.head()
```

```python
# ── Region → PnR  →  Tract → PnR ─────────────────────────────────────────────
tract_to_pnr = (
    region_to_pnr
    .merge(region_tract_weights, left_on='Origin', right_on='name', how='left')
)

tract_to_pnr['trips_allocated'] = tract_to_pnr['trips'] * tract_to_pnr['weight']

tract_to_pnr = (
    tract_to_pnr
    .groupby(['GEOID', 'Destination'])['trips_allocated']
    .sum()
    .reset_index()
    .rename(columns={'GEOID': 'tract_id', 'Destination': 'pnr', 'trips_allocated': 'trips'})
)

print(f"Tract→PnR rows  : {len(tract_to_pnr):,}")
print(f"Trips in        : {region_to_pnr['trips'].sum():,.1f}")
print(f"Trips out       : {tract_to_pnr['trips'].sum():,.1f}  (difference due to regions outside tract coverage)")
tract_to_pnr.head()
```

```python
# ── Save outputs ──────────────────────────────────────────────────────────────
out_dir = r".\Data"

pnr_to_tract.to_csv(os.path.join(out_dir, "pnr_to_tract_trips.csv"), index=False)
tract_to_pnr.to_csv(os.path.join(out_dir, "tract_to_pnr_trips.csv"), index=False)

print("Saved:")
print(f"  pnr_to_tract_trips.csv  — {len(pnr_to_tract):,} rows")
print(f"  tract_to_pnr_trips.csv  — {len(tract_to_pnr):,} rows")
```
